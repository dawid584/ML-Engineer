```python
import pandas as pd
import sklearn
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix
```

```python
SeoulBikeData = pd.read_csv(r'C:\Kodilla\SeoulBikeData\SeoulBikeData.csv', encoding= 'unicode_escape')
```

```python
SeoulBikeData 
```

```python
SeoulBikeData['Date'] = pd.to_datetime(SeoulBikeData['Date'])
```

```python
print(SeoulBikeData.info())
```

```python
SeoulBikeData = pd.get_dummies(SeoulBikeData)
SeoulBikeData.drop([ 'Hour' , 'Date' , 'Solar Radiation (MJ/m2)'], axis=1, inplace=True)
```

```python
display(SeoulBikeData)
```

```python
SeoulBikeData.describe()
```

```python
target = 'Rented Bike Count'
data = [ 'Temperature(°C)' , 'Humidity(%)' , 'Wind speed (m/s)' ,  'Visibility (10m)' , 'Dew point temperature(°C)' , 'Rainfall(mm)' , 'Snowfall (cm)' , 'Seasons_Autumn' , 'Seasons_Spring' , 'Seasons_Summer' , 'Seasons_Winter' , 'Holiday_Holiday' , 'Holiday_No Holiday' , 'Functioning Day_No' , 'Functioning Day_Yes']
```

```python
X = SeoulBikeData[data].copy()
y = SeoulBikeData[target].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

Heuristic Classifiers - RandomForestClassifier

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=1000, n_jobs=-1)
params_rf = {'max_depth': [3, 5, 10],
             'min_samples_leaf': [3, 5, 10]}
rf_gridsearch = GridSearchCV(random_forest,
                             params_rf,
                             scoring='f1_macro',
                             cv=5,
                             verbose=10, n_jobs=-1)
rf_gridsearch.fit(X_train, y_train)
print('\nBest hyperparameter:', rf_gridsearch.best_params_)
rf_model_v2 = rf_gridsearch.best_estimator_
```

```python
import tensorflow as tf
from keras.callbacks import ReduceLROnPlateau
import keras
```

```python
X = SeoulBikeData.iloc[:,1:16].values
y = SeoulBikeData.iloc[:,0].values
```

```python
print(X[0:5])
print(y[0:5])
```

```python
Y = pd.get_dummies(y).values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(2166, activation='softmax')
  ])
model
```

```python
adam_opt = keras.optimizers.Adam(learning_rate=0.005)
model.compile(optimizer=adam_opt,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2,
                              patience=5, min_lr=0.001)

model.fit(X_train, y_train, batch_size=50, epochs=100, callbacks=[reduce_lr])
```

```python
model.summary()
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```

```python
y_pred = model.predict(X_train)
```

```python
actual = np.argmax(y_train,axis=1)
predicted = np.argmax(y_pred,axis=1)
print(f"Actual: {actual}")
print(f"Predicted: {predicted}")
```

```python

fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(211)
ax.plot(actual[:200], "bp--")
ax.set_title(f'Model training', fontsize=20)
ax.set_ylabel('Training', fontsize=15)
ax.set_xlabel('Epoch', fontsize=15)
ax.legend(['Training'], loc='upper right')
        

ax = fig.add_subplot(212)
ax.plot(predicted[:200], "bp--")
ax.set_title(f'Model  Prediction', fontsize=20)
ax.set_ylabel('Prediction', fontsize=15)
ax.set_xlabel('Epoch', fontsize=15)
ax.legend(['Prediction'], loc='upper right')
plt.tight_layout()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Metrics for Neural network:')
print('Accuracy: ', accuracy_score(actual, predicted))
print("Precision:", precision_score(actual, predicted, average='weighted'))
print("Recall:", recall_score(actual, predicted, average='weighted'))
print("F1_score:", f1_score(actual, predicted, average='weighted'))
```
